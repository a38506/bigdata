```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
df = pd.read_csv('merged_data_removed_cols.csv')
```

```python
df.head()
```

# GIẢ THUYẾT TRUNG BÌNH ĐỘ PHỔ BIẾN ÂM NHẠC TĂNG DẦN ĐỀU THEO CÁC NĂM TỪ NĂM 2012 ĐẾN 2020


```python
df_grouped = df.groupby('year').size().reset_index(name='number')
filtered_df = df_grouped[(df_grouped['year'] >= 2012) & (df_grouped['year'] <= 2022)]
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered_df, x='year', y='number', palette='viridis')
plt.xticks(rotation=45)
plt.title('Số Lượng Bài Hát Theo Năm (2012-2022)')
plt.xlabel('Năm')
plt.ylabel('Số Lượng Bài Hát')
plt.show()
```

```python
filtered_df = df[(df['year'] >= 2012) & (df['year'] <= 2020)]
average_popularity = filtered_df.groupby('year')['popularity'].mean().reset_index()
```

```python
filtered_df = df[(df['year'] >= 2012) & (df['year'] <= 2020)]
average_popularity = filtered_df.groupby('year')['popularity'].mean().reset_index()
plt.pie(average_popularity['popularity'], labels=average_popularity['year'], autopct='%1.1f%%', startangle=90)
plt.title('Phần trăm Trung bình Popularity theo Năm (2012-2020)')
```

Dựa vào biểu đồ ta thấy:
    Năm có % trung bình phổ biến thấp nhất so với các năm là năm 2012 với 8.4%
    Năm có % trung bình phổ biến cao nhất so với các năm là năm 2020 với 13.6%

```python
plt.plot(average_popularity['year'], average_popularity['popularity'], marker='o', linestyle='-')
plt.xlabel('Năm')
plt.ylabel('Trung bình Popularity')
plt.title('Trung bình Popularity theo Năm (2012-2020)')
plt.show()
```

Giải thích: Do áp dụng công nghệ 4.0 vào đời sống ngày càng nhiều và số người tiếp cận với dụng nghe nhạc ngày càng tăng mặc dù số lượng bài hát ra đời qua các năm không đáng kể nhưng độ phổ biến của nó ngày càng tăng.
